```python
import pandas as pd

data = {
    'Dia': list(range(1, 31)),
    'Piezas_producidas': [980, 1000, 970, 995, 1010, 1005, 990, 985, 1020, 975,
                          980, 995, 1000, 1015, 990, 1000, 1010, 1025, 995, 980,
                          1005, 1000, 990, 970, 980, 995, 1010, 1000, 990, 985],
    'Defectos': [2, 1, 3, 2, 1, 0, 4, 2, 1, 3,
                 2, 1, 0, 1, 2, 1, 1, 0, 2, 3,
                 1, 2, 3, 4, 2, 1, 0, 1, 2, 2]
}

df = pd.DataFrame(data)
df.head()
```

```python
import matplotlib.pyplot as plt

plt.scatter(df['Piezas_producidas'], df['Defectos'], color='darkblue')
plt.title('Piezas producidas vs NÃºmero de defectos')
plt.xlabel('Piezas producidas')
plt.ylabel('Defectos')
plt.grid(True)
plt.show()
```

```python
df.plot.scatter(x='Piezas_producidas', y='Defectos', title='Piezas producidas vs Defectos')
```

```python
import seaborn as sns

sns.boxplot(y=df['Defectos'], color='lightcoral')
plt.title('DistribuciÃ³n de defectos en el mes')
plt.ylabel('NÃºmero de defectos')
plt.show()
```
