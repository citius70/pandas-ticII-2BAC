La biblioteca **Pandas** en Python tiene capacidades básicas de visualización mediante su integración con [Matplotlib](https://matplotlib.org/). 

## 6.1. Tipos de gráficos que se pueden hacer con Pandas

La función principal para graficar en Pandas es `.plot()`, que se basa en Matplotlib. Algunos de los gráficos más comunes son los que se muestran después del siguiente bloque:

````python
# Creamos el dataframe df_artistas
import pandas as pd
df_artistas = pd.read_excel("https://docs.google.com/spreadsheets/d/15pDi5KoAatuUJ1o8hcD8Lgd6zqcW_SSM/export?format=xlsx")
df_artistas

````

**Nota**: En general, en el `eje` **x** suele ir la etiqueta, y en el `eje` **y** el dato.

## - Gráfico de barras verticales

La sintaxis básica para dibujar un gráfico de barras es: `plot.bar(x,y)`

````python
df_artistas.plot.bar(
    x="Artista", 
    y="Spotify streams (millones)"
)
````

## - Grafico de barras horizontales

Sintaxis básica: `plot.barh(x,y)`

````python
df_artistas.plot.barh(
    x="Artista", 
    y="Spotify streams (millones)", 
    title="Reproducciones anuales", 
    color="green")
````

## - Gráfico de línea

Sintaxis: `plot.bar(x,y, kind='line')`

````python
df_artistas.plot(
    x='Artista',
    y='Spotify streams (millones)',
    kind='line',
    marker='x',
    color='gold',
    title="Reproducciones anuales",
    figsize=(6, 3)
)
````

## - Gráfico circular

Sintaxis básica: `plot.pie(x, y, labels, legend)`

Para este gráfico añadiremos dos argumentos

* **labels:** Este argumento especifica las etiquetas que se mostrarán.

* **legend:** Este argumento controla si se mostrará o no la leyenda en el gráfico. La leyenda generalmente proporciona información sobre el significado de los colores o las porciones en el gráfico. En este caso si la activas bloquea un poco la información.

````python
# En este caso vamos a utilizar colores elegidos por nosotros.
# Creamos un diccionario con ellos.
mis_colores = ["gold", "lightcoral", "skyblue", "palegreen", "orchid", "lightsalmon"]

# Creamos el gráfico
df_artistas.plot.pie(
    x="Artista", 
    y="Spotify streams (millones)", 
    title="Distribución de streams por artista",
    labels=df_artistas.Artista,
    autopct='%1.1f%%', # muestra porcentajes con un decimal
    colors=mis_colores,  # utilizamos los colores elegidos
    legend= False,
)
````
