Este cuaderno explora c칩mo representar datos m칠dicos de un ensayo cl칤nico utilizando gr치ficos b치sicos con `pandas`. El objetivo es analizar si un nuevo medicamento tiene mayor efectividad que un placebo, usando tres tipos de gr치ficos: **scatter plot**, **box plot** e **histograma**.

````python

import pandas as pd
import numpy as np

# Semilla para reproducibilidad
np.random.seed(0)

# Simulaci칩n de datos
n = 40
grupo = ['Tratamiento'] * 20 + ['Placebo'] * 20
edad = np.random.randint(20, 70, size=n)
dias_mejora = np.concatenate([np.random.normal(4, 1.5, 20), np.random.normal(6, 1.8, 20)]).astype(int)
intensidad = np.random.randint(5, 10, size=n)
reduccion = np.concatenate([np.random.normal(4, 1, 20), np.random.normal(2, 1, 20)])

df = pd.DataFrame({
    'Grupo': grupo,
    'Edad': edad,
    'D칤as_mejora': dias_mejora,
    'Intensidad_s칤ntomas': intensidad,
    'Reducci칩n_s칤ntomas': reduccion.round(1)
})

df.head()
````

## 游늴 Scatter plot: Edad vs D칤as hasta la mejor칤a

````python

df.plot.scatter(x='Edad', y='D칤as_mejora', title='Edad vs D칤as hasta la mejor칤a')
````

## 游닍 Box plot: Comparaci칩n entre tratamiento y placebo

````python

df.boxplot(column='Reducci칩n_s칤ntomas', by='Grupo', grid=True, figsize=(6, 5))
````

## 游늵 Histograma: Intensidad inicial de s칤ntomas

````python

df['Intensidad_s칤ntomas'].plot.hist(bins=6, title='Distribuci칩n de intensidad inicial de s칤ntomas', grid=True, edgecolor='black')
````

## 九 Conclusiones

- El scatter plot permite observar si la edad tiene relaci칩n con el tiempo de mejor칤a.
- El box plot muestra si el tratamiento tiene mayor efectividad que el placebo.
- El histograma indica c칩mo se distribuyen los s칤ntomas iniciales en la muestra.

Este tipo de an치lisis se utiliza habitualmente en medicina y ensayos cl칤nicos reales.
