Este cuaderno explora cómo representar datos médicos de un ensayo clínico utilizando gráficos básicos con `pandas`. El objetivo es analizar si un nuevo medicamento tiene mayor efectividad que un placebo, usando tres tipos de gráficos: **scatter plot**, **box plot** e **histograma**.

````python

import pandas as pd
import numpy as np

# Semilla para reproducibilidad
np.random.seed(0)

# Simulación de datos
n = 40
grupo = ['Tratamiento'] * 20 + ['Placebo'] * 20
edad = np.random.randint(20, 70, size=n)
dias_mejora = np.concatenate([np.random.normal(4, 1.5, 20), np.random.normal(6, 1.8, 20)]).astype(int)
intensidad = np.random.randint(5, 10, size=n)
reduccion = np.concatenate([np.random.normal(4, 1, 20), np.random.normal(2, 1, 20)])

df = pd.DataFrame({
    'Grupo': grupo,
    'Edad': edad,
    'Días_mejora': dias_mejora,
    'Intensidad_síntomas': intensidad,
    'Reducción_síntomas': reduccion.round(1)
})

df.head()
````

## 📉 Scatter plot: Edad vs Días hasta la mejoría

````python

df.plot.scatter(x='Edad', y='Días_mejora', title='Edad vs Días hasta la mejoría')
````

## 📦 Box plot: Comparación entre tratamiento y placebo

````python

df.boxplot(column='Reducción_síntomas', by='Grupo', grid=True, figsize=(6, 5))
````

## 📊 Histograma: Intensidad inicial de síntomas

````python

df['Intensidad_síntomas'].plot.hist(bins=6, title='Distribución de intensidad inicial de síntomas', grid=True, edgecolor='black')
````


---

## ✅ Conclusiones

- El scatter plot permite observar si la edad tiene relación con el tiempo de mejoría.
- El box plot muestra si el tratamiento tiene mayor efectividad que el placebo.
- El histograma indica cómo se distribuyen los síntomas iniciales en la muestra.

Este tipo de análisis se utiliza habitualmente en medicina y ensayos clínicos reales.
