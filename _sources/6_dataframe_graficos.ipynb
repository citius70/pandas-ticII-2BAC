# 6. Visualizacion de datos en Pandas

Aunque Pandas no es una biblioteca de visualización avanzada como [Seaborn](https://seaborn.pydata.org/) o [Plotly](https://plotly.com/), permite generar **gráficos** rápidamente para analizar datos de manera exploratoria.

La biblioteca **Pandas** en Python tiene capacidades básicas de visualización mediante su integración con [Matplotlib](https://matplotlib.org/). 

## Tipos de gráficos que se pueden hacer con Pandas:
La función principal para graficar en Pandas es `.plot()`, que se basa en Matplotlib. Algunos de los gráficos más comunes son los que se muestran después del siguiente bloque:

```python
# creamos el dataframe df_artistas
import pandas as pd
df_artistas = pd.read_excel("https://docs.google.com/uc?export=download&id=1bqVTtpAF50QSlrtB6DjfeHx1w9UKT7O3")
df_artistas.head()
```

**Nota**: En general, en el `eje` **x** suele ir la etiqueta, y en el `eje` **y** el dato.

## Gráfico de barra

La sintaxis para dibujar un gráfico de barras es: : `plot.bar(x,y)`

```python
df_artistas.plot.bar(x="Artista", y="Spotify streams (millones)")
```

## Grafico de barra horizontal

Sintaxis: `plot.barh(x,y)`

```python
df_artistas.plot.barh(x="Artista", y="Spotify streams (millones)")
```

### Gráfico de línea

Sintaxis: `plot.bar(x,y, kind='line')`

```python
df_artistas.plot(x='Artista', y='Spotify streams (millones)', kind='line', marker='o', title="Reproducciones anuales")
```

## Grafico circular

Sintaxis: `plot.pie(x, y, labels, legend)`

Para este gráfico añadiremos dos argumentos

* **labels:** Este argumento especifica las etiquetas que se mostrarán.

* **legend:** Este argumento controla si se mostrará o no la leyenda en el gráfico. La leyenda generalmente proporciona información sobre el significado de los colores o las porciones en el gráfico. En este caso si la activas bloquea un poco la información.

```python
df_artistas.plot.pie(x="Artista", y="Spotify streams (millones)", labels=df_artistas.Artista, legend= False)
```

## Grafico de linea

```python
df_dolar = pd.read_excel("https://docs.google.com/uc?export=download&id=1Uc7kM-jM0vqFlXfjdTPJSIWvniORiKIx")
df_dolar.plot.line(x="Dia", y="Valor")

# Nota: la base de datos viene con fechas en blanco, y por ende el grafico se verá "cortado".
# Para corregir eso se puede hacer con dropna() (ver sección eliminando datos vacíos)
```
