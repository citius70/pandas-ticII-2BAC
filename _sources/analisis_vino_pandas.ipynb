# üç∑ An√°lisis de Calidad de Vinos (Red Wine Dataset)

Este cuaderno usa √∫nicamente `pandas` para explorar visualmente el dataset de vinos tintos, que contiene informaci√≥n qu√≠mica de cada vino junto con su calidad num√©rica.


```python
import pandas as pd
import matplotlib.pyplot as plt

# Cargar el dataset (usa ; como separador)
url = "https://raw.githubusercontent.com/justmarkham/DAT8/master/data/winequality-red.csv"
df = pd.read_csv("datasets/winequality-red.csv")

# Mostrar primeras filas
df.head()
```

## 1. Distribuci√≥n del nivel de alcohol

```python
df['alcohol'].plot(kind='hist', bins=15, title='Distribuci√≥n de Alcohol')
plt.xlabel('Alcohol (%)')
plt.show()
```

## 2. Alcohol por nivel de calidad (Boxplot)

```python
df.boxplot(column='alcohol', by='quality')
plt.title('Alcohol por Nivel de Calidad')
plt.suptitle('')
plt.xlabel('Calidad')
plt.ylabel('Alcohol (%)')
plt.show()
```

## 3. Promedio de pH por calidad

```python
df.groupby('quality')['pH'].mean().plot(kind='bar', title='Promedio de pH por Calidad')
plt.ylabel('pH')
plt.show()
```

## 4. Relaci√≥n entre alcohol y calidad (dispersi√≥n)

```python
df.plot(x='alcohol', y='quality', kind='scatter', title='Relaci√≥n entre Alcohol y Calidad')
plt.show()
```
