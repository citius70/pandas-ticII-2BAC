# Visualizaci√≥n de datos con Pandas: Palmer Penguins üêß

En este notebook trabajar√°s con el dataset de los **ping√ºinos de Palmer**, el cual contiene datos biol√≥gicos y geogr√°ficos de diferentes especies de ping√ºinos.

Usaremos solo `pandas` para realizar las visualizaciones. ¬°A practicar!

```python
import pandas as pd

# Cargar el dataset
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/penguins.csv"
df = pd.read_csv(url)
df.head(50)
```

## Ejercicio 1: Gr√°fico de l√≠neas

Realiza un gr√°fico de l√≠neas con las columnas num√©ricas. ¬øQu√© observas?
‚û°Ô∏è Tip: Usa `df.plot()` directamente.

```python
df.plot();
```

## Ejercicio 2: Gr√°fico de barras

Agrupa los datos por especie y calcula el promedio del peso corporal (`body_mass_g`). Realiza un gr√°fico de barras con los resultados.

```python
df.groupby('species')['body_mass_g'].mean().plot(kind='bar');
```

## Ejercicio 3: Gr√°fico de pastel

Haz un gr√°fico de pastel con la cantidad de ping√ºinos por especie.

```python
df['species'].value_counts().plot(kind='pie', autopct='%1.1f%%');
```

## Ejercicio 4: Dispersi√≥n

Realiza un gr√°fico de dispersi√≥n entre `flipper_length_mm` y `body_mass_g`.

‚û°Ô∏è ¬øExiste alguna relaci√≥n?

```python
df.plot(kind='scatter', x='flipper_length_mm', y='body_mass_g');
```

## Ejercicio 5: Diagrama de caja (boxplot)

Visualiza la distribuci√≥n del largo del pico (`bill_length_mm`) por especie.

```python
df.boxplot(column='bill_length_mm', by='species');
```

## Ejercicio 6: Histograma

Crea un histograma para visualizar la distribuci√≥n de la masa corporal (`body_mass_g`).

```python
df['body_mass_g'].plot(kind='hist', bins=20);
```
